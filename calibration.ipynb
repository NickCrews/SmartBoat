```python
import pandas as pd
import json

# %matplotlib inline
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def load_data():
    with open('calibration_data.txt') as f:
        list_of_dicts = []
        for line in f:
            d = json.loads(line)
            ax, ay, az = d['acc']
            d['ax'] = ax
            d['ay'] = ay
            d['az'] = az
            mx, my, mz = d['mag']
            d['mx'] = mx
            d['my'] = my
            d['mz'] = mz
            del d['mag']
            del d['acc']
            list_of_dicts.append(d)
    df = pd.DataFrame(list_of_dicts)
    return df
```

```python
data = load_data()
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data['ax'], data['ay'], data['az'])
```
